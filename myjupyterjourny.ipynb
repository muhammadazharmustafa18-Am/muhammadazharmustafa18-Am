```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("All libraries imported successfully!")



```

```python
df = pd.read_csv('social_media_impact_on_life.csv') 
print(df.head())
```

```python


correlation_matrix = df.corr(numeric_only=True)

# 2. Setting the figure size
plt.figure(figsize=(10, 8))

# 3. Heatmap 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# 4. Title 
plt.title('Correlation Heatmap: Social Media vs Well-being Metrics')
plt.show()
```

```python

#Reg plot (Dots+ trend line)
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title('Impact of Social Media Usage on Well-being')
plt.xlabel('Average_Daily_Usage_Hours')
plt.ylabel('Sleep_Hours_Per_Night')
plt.show()
```
